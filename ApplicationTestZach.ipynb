Task 6: We might believe that the number of nurses is dependent on the number of doctors. For this we can try estimating $P(Y>2|X>1)$ and compare the result, for example to $P(Y>2|Xâ‰¤1)$ using our empirical estimate of the probability.

Compute the two probabilities below.

Task 7:
Given the data on the number of nurses ($y$) and the number of doctors ($x$),

write code that prints out the:
- marginal probability $P(X=x)$
- conditional probability $P(Y=y|X=x)$
- joint probability $P(Y=y,X=x)$

````python
num_doctors=1
num_nurses=2
### TODO: your answer here

# Drop rows with missing values in either column for consistent calculations
filtered_data = data.dropna(subset=['num_doctors_fulltime', 'num_nurses_fulltime'])

# Calculate marginal probability P(X=x)
total_facilities = filtered_data.shape[0]
facilities_with_x_doctors = filtered_data[filtered_data['num_doctors_fulltime'] == num_doctors].shape[0]
p_x = facilities_with_x_doctors / total_facilities if total_facilities > 0 else 0

# Calculate joint probability P(Y=y, X=x)
facilities_with_x_doctors_and_y_nurses = filtered_data[(filtered_data['num_doctors_fulltime'] == num_doctors) & (filtered_data['num_nurses_fulltime'] == num_nurses)].shape[0]
p_y_and_x = facilities_with_x_doctors_and_y_nurses / total_facilities if total_facilities > 0 else 0

# Calculate conditional probability P(Y=y|X=x)
# Using the formula P(Y=y|X=x) = P(Y=y, X=x) / P(X=x)
p_y_given_x = p_y_and_x / p_x if p_x > 0 else 0
````

````python
print("P(x) is", p_x)
print("P(y|x) is", p_y_given_x)
print("P(y,x) is", p_y_and_x)
````

Task 8:
Verify the Sum Rule by writing code that computes $P(Y=y)$ by adding $P(P=y, X=x)$ for all values of $x$.

````python
### TODO: your answer here
# We'll use the filtered_data from the previous task for consistency
# num_nurses is already defined as y

# Calculate P(Y=y) directly
facilities_with_y_nurses = filtered_data[filtered_data['num_nurses_fulltime'] == num_nurses].shape[0]
p_y_direct = facilities_with_y_nurses / total_facilities if total_facilities > 0 else 0

# Calculate P(Y=y) by summing P(Y=y, X=x) over all x
sum_of_joint_probabilities = 0
unique_doctors = filtered_data['num_doctors_fulltime'].unique()

for doctor_count in unique_doctors:
    # Calculate joint probability P(Y=y, X=x) for the current x (doctor_count)
    facilities_with_x_doctors_and_y_nurses_sum = filtered_data[(filtered_data['num_doctors_fulltime'] == doctor_count) & (filtered_data['num_nurses_fulltime'] == num_nurses)].shape[0]
    p_y_and_x_current = facilities_with_x_doctors_and_y_nurses_sum / total_facilities if total_facilities > 0 else 0
    sum_of_joint_probabilities += p_y_and_x_current

print(f"Marginal probability P(Y={num_nurses}) calculated directly: {p_y_direct}")
print(f"Marginal probability P(Y={num_nurses}) calculated by summing joint probabilities: {sum_of_joint_probabilities}")
````
